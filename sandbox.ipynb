```python
import pandas as pd

dfœ = pd.read_csv("Olist.csv")
int_col = ["payment_sequential","payment_installments","order_item_id","review_score","review_answer_timestamp","product_photos_qty","order_delivered_customer_date"]
object_col = ["order_item_id","customer_zip_code_prefix"]

for col in int_col:

    dfœ[col] = pd.to_numeric(dfœ[col], errors='coerce',downcast='integer')

for col in object_col:
    dfœ[col] = dfœ[col].astype(str)
df = dfœ[["customer_unique_id","payment_value","order_purchase_timestamp","customer_city"]]

dfœ.info()

```

```python
from statistics import mode


def compute_mode(x):
    try:
        return mode(x)
    except:
        return None
        
result = df.groupby('customer_unique_id').agg({'payment_value': 'mean','order_purchase_timestamp':'min', 'customer_city': lambda x: compute_mode(x)})

result
```

```python
mode_col = ['payment_sequential','payment_type','payment_installments','product_id','seller_state',
            'seller_id','customer_zip_code_prefix','customer_city','product_category_name','customer_state','seller_zip_code_prefix','seller_city']
mean_col = ['payment_value','price','freight_value','review_score']
last_col = ['order_purchase_timestamp','order_delivered_customer_date','order_estimated_delivery_date','review_answer_timestamp']
count_col = ['order_id']

agg_dict = {}

for col in mode_col:
    agg_dict[col] = lambda x: compute_mode(x)

for col in mean_col:
    agg_dict[col] = 'mean'

for col in last_col:
    agg_dict[col] = 'min'

for col in count_col:
    agg_dict[col] = 'count'
```

```python
result = dfœ.groupby('customer_unique_id').agg(agg_dict)
```
