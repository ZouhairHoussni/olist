## Manipulation

```python
import pandas as pd

df = pd.read_csv("Olist.csv")

# cette transformation est juste pour avoir int8 au lieu de int64
int_col = ["payment_sequential","payment_installments","order_item_id","review_score","review_answer_timestamp","product_photos_qty","order_delivered_customer_date"]
object_col = ["order_item_id","customer_zip_code_prefix"]

for col in int_col:

    df[col] = pd.to_numeric(df[col], errors='coerce',downcast='integer')

for col in object_col:
    df[col] = df[col].astype(str)
    
df.head()
```

```python
df[df["payment_sequential"].isna()]
```

## Group by & AggrÃ©gation 

```python
# SÃ©paration des colonnes par mÃ©thode d'aggrÃ©gation

mode_col = ['payment_sequential','payment_type','payment_installments','product_id','seller_state',
            'seller_id','customer_zip_code_prefix','customer_city','product_category_name','customer_state','seller_zip_code_prefix','seller_city','product_photos_qty']
mean_col = ['payment_value','price','freight_value','review_score']
last_col = ['order_purchase_timestamp','order_delivered_customer_date','order_estimated_delivery_date','review_answer_timestamp']
count_col = ['order_id']
```

```python
from statistics import mode

# crÃ©ation d'une fonction qui fait le mode

def compute_mode(x):
    try:
        return mode(x)
    except:
        return None
```

```python
# crÃ©ation d'un dictionnaire qui lie la colonne Ã  la mÃ©thode choisit dans la fonction d'aggrÃ©gation

agg_dict = {}

for col in mode_col:
    agg_dict[col] = lambda x: compute_mode(x)

for col in mean_col:
    agg_dict[col] = 'mean'

for col in last_col:
    agg_dict[col] = 'min'

for col in count_col:
    agg_dict[col] = 'count'
```

```python
# application de l'aggrÃ©gation

df_agg = df.groupby('customer_unique_id').agg(agg_dict)

df_agg
```

## R.F.M score

### RÃ©cence

```python
import numpy as np

# DÃ©finition les point Ã  accorder
pts = [15,10,6,3,0]

# DÃ©finition des conditions
conditions = [
    df_agg['order_purchase_timestamp'] < 93,
    (df_agg['order_purchase_timestamp'] >= 93) & (df_agg['order_purchase_timestamp'] < 185),
    (df_agg['order_purchase_timestamp'] >= 185) & (df_agg['order_purchase_timestamp'] < 276),
    (df_agg['order_purchase_timestamp'] >= 276) & (df_agg['order_purchase_timestamp'] < 366),
    df_agg['order_purchase_timestamp'] >= 366
]

# Application et crÃ©ation de la nouvelle colonne 'R' 
df_agg["R"] = np.select(conditions,pts)
```

### FrÃ©quence

```python
# selection des donnÃ©es des 12 derniers mois
df_last_year = df[df["order_purchase_timestamp"]<366]

df_last_year
```

```python
# aggrÃ©gation par nombre de fois de commande par client

df_year_agg = df_last_year.groupby('customer_unique_id')['order_id'].nunique().reset_index(name='unique_orders_count')

df_year_agg
```

```python
# application du bonnus de 3 points pour chaque commande 

df_year_agg["unique_orders_count"] = df_year_agg["unique_orders_count"].apply(lambda x : x*3)

# changement du nom de la colonne en 'F' pour frÃ©quence

df_year_agg = df_year_agg.rename(columns={'unique_orders_count':'F'})

df_year_agg
```

```python
# merger le dataframe gÃ©nÃ©ral avec le rÃ©sultant en left, en remplanÃ§ant les nan par 0

df_agg = df_agg.merge(df_year_agg,how='left',on='customer_unique_id').fillna(0)

# petite transformation en int8

df_agg["F"] = pd.to_numeric(df_agg["F"], errors='coerce',downcast='integer')

df_agg
```

### Montant

```python
# groupement par client en aggrÃ©geant le payment en somme 

df_payment_agg = df.groupby('customer_unique_id')['payment_value'].sum().reset_index(name='M')

# transformer 2% du montant en pts, Ã§Ã  revient Ã  diviser par 50 le montant, et arrondir au nombre entier 
df_payment_agg['M'] = df_payment_agg['M'].apply(lambda x : int(round(x/50)))

# maximiser le nombre de points au 10 pts max
df_payment_agg['M'] = df_payment_agg['M'].apply(lambda x : 10 if x > 10 else x)

df_payment_agg
```

```python
# merger le dataframe gÃ©nÃ©ral avec le rÃ©sultant en left, en remplanÃ§ant les nan par 0

df_agg = df_agg.merge(df_payment_agg,how='left',on='customer_unique_id')

df_agg
```

```python
# create an RFM score from the 3 other columns

df_agg['RFM'] = df_agg['R'] + df_agg['F'] + df_agg['M']

# save the dataframe as csv file

df_agg.to_csv("Olist_cleaned.csv",index=False)
```
