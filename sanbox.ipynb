```python
import pandas as pd

dfœ = pd.read_csv("Olist.csv")

df = dfœ[["customer_unique_id","payment_value","order_purchase_timestamp","order_item_id"]]

df

```

```python
from statistics import mode


def compute_mode(x):
    try:
        return mode(x)
    except:
        return None
result = df.groupby('customer_unique_id').agg({'payment_value': 'mean', 'order_item_id': lambda x: compute_mode(x)})
```
