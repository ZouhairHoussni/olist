```python
import pandas as pd

df = pd.read_csv('Olist.csv')

df.head()
```

```python
df_a = df[['order_purchase_timestamp','order_estimated_delivery_date','order_delivered_customer_date','review_score']]

df_a
```

```python
df_a['jour_livraison'] = df_a['order_purchase_timestamp'] - df_a['order_delivered_customer_date']

df_a['delay'] = df_a['order_estimated_delivery_date'] - df_a['order_delivered_customer_date']
df_a = df_a[df_a['delay']<0]
df_b = df_a.groupby('review_score').size().reset_index()

df_b
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lmplot(data=df_a,x='jour_livraison',y='review_score',line_kws={'color': 'red'})
```

```python
from scipy.stats import chi2_contingency
import numpy as np

contingency = pd.crosstab(df_a['review_score'], np.sign(df_a['jour_livraison']))

# Calculate the chi-squared test statistic and associated p-value
chi2, p, dof, expected = chi2_contingency(contingency)

# Calculate Cramér's V
n = contingency.sum().sum()
V = np.sqrt(chi2 / (n * (min(contingency.shape)-1)))

# Print the correlation
print('Correlation between review_score and delivery_delay:', V)
```
